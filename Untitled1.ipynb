```python
import openpyxl
import urllib.request
from datetime import datetime
def getdata(place):
  # Specify the URL of the text file
  url = 'https://avdc.gsfc.nasa.gov/pub/data/satellite/Aura/OMI/V03/L2OVP/OMUVB/aura_omi_l2ovp_omuvb_v03_'+place+'.txt'

  # Download the file
  urllib.request.urlretrieve(url, 'aura_omi_l2ovp_omuvb_v03_kathmandu.txt')

  # Open the text file in read mode
  with open('aura_omi_l2ovp_omuvb_v03_kathmandu.txt', 'r') as file:
    # Read the contents of the file into a list of lines
    lines = file.readlines()

  # Create a new workbook
  workbook = openpyxl.Workbook()

  # Get the active worksheet
  worksheet = workbook.active

  lines[50] = lines[50].replace("sec. (UT)", "sec.(UT)")

  # Split the 51st line into a list of values
  heading = lines[50].split()

  # Write the values to the first row of the worksheet
  for i, value in enumerate(heading):
    worksheet.cell(row=1, column=i+1).value = value

  # Iterate over the lines of data starting at line 52
  for i, line in enumerate(lines[51:]):
    # Split the line into a list of values
    values = line.split()

    # Iterate over the values
    for j, value in enumerate(values):
      # Write the value to the cell
      if j == 0:
        worksheet.cell(row=i+2, column=j+1).value = datetime.strptime(value, '%Y%m%dT%H%M%S%fZ')
      else:
        worksheet.cell(row=i+2, column=j+1).value = float(value)

  # Save the workbook
  workbook.save(f'{place}.xlsx')
  return workbook
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create lists to store the data
dates = []
ird305 = []
ird310 = []
ird324 = []
ird380 = []

data = worksheet.values
columns = next(data)[0:]

df = pd.DataFrame(data,columns=columns)
print(df.loc[:,'Datetime'])
# Iterate over each row in the worksheet
for row in worksheet.iter_rows():
  # Get the values of the cells

  date = datetime.strptime(row[0].value, "%m/%d/%Y %I:%M:%S %p")

  ird305_value = row[26].value
  ird310_value = row[27].value
  ird324_value = row[28].value
  ird380_value = row[29].value

  # Append the values to the lists
  dates.append(date)
  ird305.append(ird305_value)
  ird310.append(ird310_value)
  ird324.append(ird324_value)
  ird380.append(ird380_value)

# Get the years of the data
years = set([date.year for date in dates])

# Iterate over the years
for year in years:
  # Create a figure
  fig, ax = plt.subplots()

  # Get the data for the year
  year_dates = [date for date in dates if date.year == year]
  year_ird305 = [ird305[i] for i in range(len(dates)) if dates[i].year == year]
  year_ird310 = [ird310[i] for i in range(len(dates)) if dates[i].year == year]
  year_ird324 = [ird324[i] for i in range(len(dates)) if dates[i].year == year]
  year_ird380 = [ird380[i] for i in range(len(dates)) if dates[i].year == year]

  # Plot the data
  ax.plot(year_dates, year_ird305, label='IRD305')
  ax.plot(year_dates, year_ird310, label='IRD310')
  ax.plot(year_dates, year_ird324, label='IRD324')
  ax.plot(year_dates, year_ird380, label='IRD380')

  # Add a legend
  ax.legend()

  # Show the plot
  plt.show()
```
