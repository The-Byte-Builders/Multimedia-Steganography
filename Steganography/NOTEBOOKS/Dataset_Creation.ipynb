```python
import os
import random
import shutil
import pandas as pd

# Set the path to the folder containing the images
image_path = '/content/drive/MyDrive/Colab Notebooks/DatasetPool256'

# Set the path to the destination folders
secret_path = '/content/drive/MyDrive/Colab Notebooks/Dataset256/secret'
cover_path = '/content/drive/MyDrive/Colab Notebooks/Dataset256/cover'

# Get a list of all the images in the folder
images = os.listdir(image_path)
print(len(images))
images = images[:2000]

# Randomly shuffle the list of images
random.shuffle(images)

# Split the images list into two sub-lists
half = len(images) // 2
secret_images = images[:half]
cover_images = images[half:]

# Move the secret images to the secret folder
secret_img_list = []
for image_name in secret_images:
    source_path = os.path.join(image_path, image_name)
    destination_path = os.path.join(secret_path, image_name)
    shutil.move(source_path, destination_path)
    secret_img_list.append(image_name)

# Move the cover images to the cover folder
cover_img_list = []
for image_name in cover_images:
    source_path = os.path.join(image_path, image_name)
    destination_path = os.path.join(cover_path, image_name)
    shutil.move(source_path, destination_path)
    cover_img_list.append(image_name)

# Create a dataframe from the cover and secret image lists
dataset = pd.DataFrame(columns=['cover_image', 'secret_image'])
for i in range(len(cover_img_list)):
    dataset = dataset.append({
        'cover_image': cover_img_list[i],
        'secret_image': secret_img_list[i]
    }, ignore_index=True)

# Save the dataframe as a CSV file
dataset.to_csv('/content/drive/MyDrive/Colab Notebooks/Dataset_x256.csv', index=False)
```
